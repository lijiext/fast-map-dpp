```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
import matplotlib
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
# pandas 可以读取 json 和保存 json，也可以获取 api 的数据
dpp_res = pd.read_json('dpp_res_array.json')
```

```python
# 筛选出 dpp_res 中 n == 100, d == 3 的数据
data = dpp_res[(dpp_res['n'] == 100)]
data
```

```python
# sns.set(style="ticks", color_codes=True)
sns.set_theme(style="whitegrid",font='Times New Roman', rc={"lines.linewidth": 1.5, 'grid.linestyle': '--'})
sns.set_palette("winter", 8, 0.5)
sns.set_context(context='paper', font_scale=1)
plt.figure(dpi=300)
plt.title('RMDSE of DPP N=1000')
plt.xlabel('TopK')
plt.ylabel('RMDSE')
# plt.legend(labels=[3, 4, 5, 6, 7, 8],title='Dimensions')
sns.lineplot(x='k', y='rmdse',hue= 'd',  data=data, palette="muted")
# legend = plt.legend()  
```
